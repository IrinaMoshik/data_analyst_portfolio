```python
import pandas as pd
import openpyxl
```

```python
"""Завантаження таблиць"""
df_apr=pd.read_excel('Регистр Лиды с CRM за Апрель 18.xlsx')
df_mar=pd.read_excel('Регистр Лиды с CRM за Март 18.xlsx')
df_jun=pd.read_excel('Регистр Лиды с CRM за Июнь 18.xlsx')
df_jul=pd.read_excel('Регистр Лиды с CRM за Июль 18.xlsx')
df_may=pd.read_excel('Регистр Лиды с CRM за Май 18.xlsx')
```

```python
df_mar.head()
```

```python
"""Обьєднання всіх таблиць в один датафрейм"""
combined_df = pd.concat([df_mar, df_apr, df_may, df_jun, df_jul], ignore_index=True)
```

```python
combined_df.head()
```

```python
"""Перевірка формату даних"""
combined_df.info()
```

```python
"""Міняю формат колонки 'Период'"""
combined_df['Период']=pd.to_datetime(combined_df['Период'], format='%d.%m.%Y')
```

```python
"""Відфільтровую строки, де "Менеджер" не зазначено, тобто лід не був взятий в роботу"""
converted_leads = combined_df.dropna(subset=['Менеджер'])
converted_leads.head()
```

```python
"""Виберу найменшу дату 'Періоду' для кожного ліда, щоб зрозуміти коли він був взятий в роботу"""
min_period_by_lead = converted_leads.groupby('Номер лида')['Период'].min().dt.month.reset_index()
min_period_by_lead=min_period_by_lead.rename(columns={'Период': 'Взятий в роботу'})
min_period_by_lead
```

```python
"""Обьєднаю дані про найменшу дату з основним DataFrame, лівий джоін, бо нам потрібні всі дані з основної таблички"""
converted_leads2 = pd.merge(converted_leads, min_period_by_lead, on=['Номер лида'], how='left')
converted_leads2.head()
```

```python
"""Порахую конверсії, сгруповані по місяцях і по менеджерах"""

leads_count2 = converted_leads2.groupby(['Взятий в роботу', 'Менеджер'])['Номер лида'].nunique().reset_index()
leads_count2.columns = ['Місяць', 'Менеджер', 'Кількість_лідів']

payments_count2 = converted_leads2[converted_leads2['Результат звонка'] == 'Оплата'].groupby(['Місяць', 'Менеджер']).size().reset_index()
payments_count2.columns = ['Місяць', 'Менеджер', 'Кількість_оплат']

```

```python
"""Обьєдную в один датафрейм"""
monthly_conversion_by_manager = pd.merge(leads_count2, payments_count2, on=['Місяць', 'Менеджер'], how='inner')
monthly_conversion_by_manager
```

```python
"""Розраховую і додаю колонку конверсії"""
monthly_conversion_by_manager['Конверсія'] = round((monthly_conversion_by_manager['Кількість_оплат'] / monthly_conversion_by_manager['Кількість_лідів']) * 100, 1)

monthly_conversion_by_manager
```

```python
"""Збережу табличку для візуалізації"""
monthly_conversion_by_manager.to_csv('monthly_conversion_by_manager.csv')
```

```python
"""Порахую кількість часу, скільки проходить від взяття в роботу до оплати"""


min_period_by_l = converted_leads.groupby('Номер лида')['Период'].min().reset_index()
min_period_by_l=min_period_by_l.rename(columns={'Период': 'Дата взяття в роботу'})


"""Обьєднаю дані про найменшу дату з основним DataFrame, лівий джоін, бо нам потрібні всі дані з основної таблички"""
converted_leads3 = pd.merge(converted_leads, min_period_by_l, on=['Номер лида'], how='left')
converted_leads3.head()

```

```python
"""Рахую кількість днів між оплатою і взяттям в роботу"""
converted_leads3['Difference'] = converted_leads3[converted_leads3['Результат звонка'] == 'Оплата']['Период'] -converted_leads3[converted_leads3['Результат звонка'] == 'Оплата']['Дата взяття в роботу']
```

```python
max = round(converted_leads3['Difference'].dt.days.max())
mean = round(converted_leads3['Difference'].dt.days.mean())
median = round(converted_leads3['Difference'].dt.days.median())
print (f'max= {max} mean= {mean} median={median}')
```

```python
days_before_purchase=converted_leads3['Difference'].value_counts().reset_index()
days_before_purchase.to_csv('days_before_purchase.csv')
days_before_purchase
```

```python
"""Порахую конверсії, сгруповані по місяцях і по менеджерах"""

leads_count_new = converted_leads2.groupby(['Взятий в роботу', 'Менеджер', 'Продукт'])['Номер лида'].nunique().reset_index()
leads_count_new.columns = ['Місяць', 'Менеджер', 'Продукт', 'Кількість_лідів']

payments_count_new = converted_leads2[converted_leads2['Результат звонка'] == 'Оплата'].groupby(['Місяць', 'Менеджер', 'Продукт']).size().reset_index()
payments_count_new.columns = ['Місяць', 'Менеджер', 'Продукт', 'Кількість_оплат']
```

```python
"""Обьєдную в один датафрейм"""
monthly_conversion_by_manager_product = pd.merge(leads_count_new, payments_count_new, on=['Місяць', 'Менеджер', 'Продукт'], how='inner')
monthly_conversion_by_manager_product['Конверсія'] = round((monthly_conversion_by_manager_product['Кількість_оплат'] / monthly_conversion_by_manager_product['Кількість_лідів']) * 100, 1)


monthly_conversion_by_manager_product
```
